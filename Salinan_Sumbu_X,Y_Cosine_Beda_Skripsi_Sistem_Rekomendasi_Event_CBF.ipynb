### Word Extraction

```python
# event['nama_event'] = event['nama_event'].str.lower()
# event['tags'] = event['tags'].str.lower()
```

```python
event_keys['nama_event'] = event_keys['nama_event'].astype(str).str.replace('\d+', '').str.replace('(', '').str.replace(')', '').str.replace('.', '')
event_keys['tags'] = event_keys['tags'].astype(str).str.replace('\d+', '').str.replace('#', '').str.replace('_', '')
event_keys
```

```python
# event['nama_event2'] = event['nama_event'].apply(word_tokenize)
# event['keywords'] = event['tags'].apply(word_tokenize)
```

```python
# fungsi untuk keyword
def get_keywords(x):
    plot = x
    # inisialisasi Rake menggunakan stopword bahasa Inggris dari NLTK, dan semua karakter tanda baca
    r = Rake()
    # extract keywords from text
    r.extract_keywords_from_text(plot)
    # get dictionary with keywords and scores
    scores = r.get_word_degrees()
    # return new keywords as list, ignoring scores
    return(list(scores.keys()))

# Apply function to generate keywords
event_keys['tags2'] = event_keys['tags'].apply(get_keywords)
event_keys['jenis_event2'] = event_keys['jenis_event'].apply(get_keywords)
event_keys['nama_event2'] = event_keys['nama_event'].apply(get_keywords)
event_keys
```

```python
list_stopwords = set(stopwords.words('indonesian'))
def stopwords_removal(words):
    return [word for word in words if word not in list_stopwords]

event_keys['nama_event2'] = event_keys['nama_event2'].apply(stopwords_removal)
event_keys['tags2'] = event_keys['tags2'].apply(stopwords_removal)
event_keys
```

```python
import time

start = time.time()
factory = StemmerFactory()
stemmer = factory.create_stemmer()

def stemming(word):
  # token without stopword
  list_tokens = word
  return [(stemmer.stem(str(token))) for token in list_tokens]

# stem
event_keys['nama_event2'] = event_keys['nama_event2'].apply(stemming)

end = time.time()
print(end - start)
event_keys
```

```python
event_keys.set_index('event_id', inplace = True)
event_keys.head()
```

```python
# fungsi bag_word
event_bow = pd.DataFrame() 

# event_keys['nama_event'] = event['nama_event']
# event_keys['bag_of_words'] = ''

def bag_words(x):
    return(' '.join(x['nama_event2'])+ ' ' +' '.join(x['tags2']))
event_bow['bag_of_words'] = event_keys.apply(bag_words, axis = 1)

event_bow
# + ' ' +''.join(x['jenis_event'])
```

```python
indices = pd.Series(event_bow.index)

def recommendations(nama):
    
    recommended = []
    
    # gettin the index that matches the nama
    idx = indices[indices == nama].index[0]
    event_bow_idx = event_bow[idx:idx+1]

    tfidf = TfidfVectorizer(min_df=1, norm=None, smooth_idf=False, stop_words='english')
    tfidf_matrix = tfidf.fit_transform(event_bow['bag_of_words'])
    tf_idf_q = tfidf.transform(event_bow_idx['bag_of_words'])

    cosine_sim = cosine_similarity(tf_idf_q, tfidf_matrix)
    # df_cosine_sim = pd.DataFrame(cosine_sim, index=event.index, columns=event.index)
    # creating a Series with the similarity scores in descending order
    score_series = pd.Series(cosine_sim[0]).sort_values(ascending = False)
    top_indexes = list(score_series.iloc[1:11].index)
    event['cosine_score'] = score_series
    
    # for i in top_indexes:
    #     recommended.append(list(event['nama_event'])[i])
    
    return pd.DataFrame(event[['event_id', 'nama_event', 'cosine_score']].iloc[top_indexes])
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
from sklearn import svm
# from sklearn.model_selection import cross_val_score

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)

X_train = tfidf.fit_transform(X_train)
X_test = tfidf.transform(X_test)

print('Train Data: ', X_train.shape)
print('Test Data: ', X_test.shape)

clf = svm.SVC()
clf.fit(X_train, y_train)
predict = clf.predict(X_test)

print(metrics.classification_report(y_test, predict,
                                            target_names=['Beasiswa','Festival-Konser','Lomba','Talkshow','Webinar','Workshop']))
```

```python
# Import confusion matrix dari sklearn
from sklearn.metrics import confusion_matrix
import seaborn as sns

def display_conf(y_test,predict):
    cm = sns.heatmap(confusion_matrix(y_test,predict),annot=True,cbar=False, fmt="d", cmap='Blues')
    cm.set_title('Confusion Matrix')
    cm.set_ylabel('Actual')
    cm.set_xlabel('Prediction')
    
    cm.yaxis.set_ticklabels(['Beasiswa', 'Festival-Konser', 'Lomba', 'Talkshow', 'Webinar', 'Workshop'], rotation=0)
    cm.xaxis.set_ticklabels(['Beasiswa', 'Festival-Konser', 'Lomba', 'Talkshow', 'Webinar', 'Workshop'])
    
    plt.show()

# Memanggil fungsi untuk menampilkan visualisasi confusion matrix
display_conf(y_test,predict)
```

```python
# findf_tfidf = TfidfVectorizer()
from sklearn.naive_bayes import MultinomialNB

X_train, X_test, y_train, y_test = train_test_split(x, y, 
                                                    test_size=0.3)
X_train_NB = tfidf.fit_transform(X_train)
X_test_NB = tfidf.transform(X_test)

# print(X_train_NB.shape)
# print(X_test_NB.shape)

naive_bayes_classifier = MultinomialNB()
naive_bayes_classifier.fit(X_train_NB,y_train)

y_pred_NB = naive_bayes_classifier.predict(X_test_NB)

print(metrics.classification_report(y_test, y_pred_NB,
                                            target_names=['Beasiswa','Festival-Konser','Lomba','Talkshow','Webinar','Workshop']))
```

```python
# Import confusion matrix dari sklearn
from sklearn.metrics import confusion_matrix
import seaborn as sns

def display_conf(y_test,y_pred_NB):
    sns.heatmap(confusion_matrix(y_test,y_pred_NB),annot=True,linewidths=3,cbar=False, fmt="d")
    plt.title('Confusion Matrix')
    plt.ylabel('Actual')
    plt.xlabel('Prediction')
    plt.show()

# Memanggil fungsi untuk menampilkan visualisasi confusion matrix
display_conf(y_test,y_pred_NB)
```

```python
from sklearn.neighbors import KNeighborsClassifier

X_train, X_test, y_train, y_test = train_test_split(x, y, 
                                                    test_size=0.3)

knn = KNeighborsClassifier(n_neighbors=12)

X_train = tfidf.fit_transform(X_train)
X_test = tfidf.transform(X_test)

knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
print(metrics.classification_report(y_test, y_pred, target_names=['Beasiswa','Festival-Konser','Lomba','Talkshow','Webinar','Workshop']))
```

```python
# try K=1 through K=30 and record testing accuracy
k_range = list(range(1, 31))
scores = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    scores.append(metrics.accuracy_score(y_test, y_pred))
```

```python
# import Matplotlib (scientific plotting library)
import matplotlib.pyplot as plt
# allow plots to appear within the notebook
%matplotlib inline
# plot the relationship between K and testing accuracy
plt.plot(k_range, scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Testing Accuracy')
```
